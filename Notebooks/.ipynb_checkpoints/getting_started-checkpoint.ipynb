```python
def clean_daily(df):
    # Split date into day, month, year
    days = [date.day for date in df.index][:-1]
    months = [date.month for date in df.index][:-1]
    years = [date.year for date in df.index][:-1]
    
    # Get all the other columns
    daily_open = df["1. open"].tolist()[:-1]
    daily_high = df["2. high"].tolist()[:-1]
    daily_low = df["3. low"].tolist()[:-1]
    daily_close = df["4. close"].tolist()[:-1]
    daily_volume = df["5. volume"].tolist()[:-1]
    
    # TARGET === next day's close
    next_close = df["4. close"].tolist()[1:]
    
    # Send to new df
    clean_df = pd.DataFrame(data = {
        "day": days, 
        "month": months, 
        "year": years, 
        "open": daily_open, 
        "high": daily_high, 
        "low": daily_low, 
        "close": daily_close, 
        "volume": daily_volume, 
        "next_close": next_close
    })
    
    # Return clean_df
    return clean_df
```
